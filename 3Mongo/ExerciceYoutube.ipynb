```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient('mongo')
database = client['exercices']
collection = database['youtube']
```

```python
from datetime import datetime
df_youtube = pd.read_csv("./data/USvideos.csv", parse_dates=['date'])
df_youtube['date'] = df_youtube['date'] + '.2020'
df_youtube['date'] = df_youtube['date'].apply(lambda x : datetime.strptime(x, "%d.%M.%Y"))
df_youtube['tags'] = df_youtube['tags'].apply(lambda x : x.split('|'))
df_youtube.head()
```

## Question 0

### Netoyer les données

### Importer les données

```python
import json
with open("./data/US_category_id.json",'r') as load_f:
    json_youtube = json.load(load_f)
    for element in json_youtube['items'] : 
        element['channelId'] = element['snippet']['channelId']
        element['title'] = element['snippet']['title']
        element['assignable'] = element['snippet']['assignable']
        element.pop('snippet')
    json_youtube = pd.DataFrame.from_dict(json_youtube['items'])
json_youtube[['id']] = json_youtube[['id']].astype(int)
json_youtube.head()
```

```python
df = df_youtube.merge(json_youtube,  left_on ='category_id', right_on = 'id')
df[df['channel_title']=="Apple"]
```

```python
df.isnull().sum()
```

```python
dict_donnees = df.to_dict('records')
```

```python
collection.insert_many(dict_donnees)
```
